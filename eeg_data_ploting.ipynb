## ploting first 1000 samples

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data
data = pd.read_csv("eeg_train.csv")

# Plot the first 1000 data points for each column
for column in data.columns[:-1]:  # Exclude the last column (label)
    plt.figure(figsize=(10, 5))
    plt.plot(data[column][:1000])
    plt.title(column)
    plt.xlabel("Data Point")
    plt.ylabel("Value")
    plt.show()
```

## plotting eye blinking point in graph

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV data
data = pd.read_csv("eeg_train.csv")

# Plot the first 1000 data points for each column
for column in data.columns[:-1]:  # Exclude the last column (label)
    plt.figure(figsize=(10, 5))
    
    # Plot all data points
    plt.plot(data[column][:10000], label="All Data")

    # Highlight points where label is 1 with red dots
    label_1_indices = data[data['label'] == 1].index
    plt.scatter(label_1_indices.intersection(range(10000)), data[column].iloc[label_1_indices.intersection(range(10000))], color='red', label="Label 1")

    plt.title(column)
    plt.xlabel("Data Point")
    plt.ylabel("Value")
    plt.legend()
    plt.show()
```

## find correlation of 2 users using 14 channels

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV data for User 1
user1_df = pd.read_csv("eeg_train.csv") #Certainly! Here's a revised version of your sentence:

#"Please note that we are utilizing a single CSV file. If you wish to obtain correlations between two users, 
# please use separate CSV files for each user."

# Read the CSV data for User 2
user2_df = pd.read_csv("eeg_train.csv")

# Selecting the first 100 data rows and the first 14 data columns
selected_data_1 = user1_df.iloc[:1000, :14]

# Selecting the first 100 data rows and the first 14 data columns
selected_data_2 = user2_df.iloc[1000:2000, :14]

# Concatenate the data for both users along rows
combined_df = pd.concat([selected_data_1, selected_data_2], ignore_index=True)

# Selecting the first 14 data columns for each user
selected_columns = combined_df.columns[:14]

# Calculate the correlation matrix for both users
combined_correlation_matrix = combined_df[selected_columns].corr()

# Plot the correlation matrix
plt.figure(figsize=(8, 8))
sns.heatmap(combined_correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=.5)
plt.title("Combined Users - Correlation Matrix of EEG Data Columns")
plt.show()
```
